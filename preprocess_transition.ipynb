<h2>Get CT Slice Chunks</h2>

```python
#load scans from .npy files
import os
import numpy as np

#extract chunk number from filename
def get_chunk_num(chunk):
  num = int(chunk[5:-4])
  return(num)

#get scan chunks
os.chdir('drive/My Drive/mid_slices')
print(os.getcwd())
scan_chunks = {}
for (dir,root,files) in os.walk(os.getcwd()):
  for file in files:
    #if file ends in extenstion '.npy' then add chunk to dict
    if '.npy' in file:
      num = get_chunk_num(file)
      scan_chunks[num] = file

#sort scan chunks
sorted_chunks = sorted(scan_chunks.keys())
print(f'length of scan chunks: {len(sorted_chunks)}')






  
```

```python
import time
import cv2
from numpy import save, asarray
from keras.preprocessing.image import array_to_img,save_img
import tensorflow as tf
from time import sleep

#start counter
start = time.perf_counter()

#resize image to (224,224)
def decode_img(img,IMG_WIDTH,IMG_HEIGHT):
  return(cv2.resize(img,(IMG_WIDTH,IMG_HEIGHT)))

#initialize chunk count and data array
chunk_count = 0
data = []

#load scan chunks
for chunk_ in sorted_chunks:

  #load chunk
  chunk = np.load(scan_chunks[chunk_])

  #reset first var
  first = True

  #add each scan to data array and convert to (224,224,3) format
  for scan in chunk:
    #convert to (224,224) format
    scan = decode_img(scan,224,224)

    #if scan is not in 3 channel format then convert to 3 channels
    if scan.ndim < 3:
      scan = np.asarray([scan]*3)
      scan = scan.reshape(224,224,3)


    #append scan to data array
    data.append(scan)

    #if we are handling first scan in chunk, display information
    if first:
      print(f'{scan_chunks[chunk_]} loaded at: {time.perf_counter()}')
      print(f'data length: {len(data)}')
      print("\n")
      first = False
      chunk_count += 1

    
  
#finish counter and process time
finish = time.perf_counter()
print(f'this process took {finish-start} seconds')


```

<h2>Load Pretrained InceptionV3 base</h2>

```python
#import inceptionv3 model with imagenet weights 
from keras.applications import InceptionV3 ##

#define input image shape
img_shape = (224,224,3)

#load inceptionv3
inception = InceptionV3(weights='imagenet', include_top=False, input_shape=img_shape)





    
```

<h2> Extract features of Transition data with InceptionV3 base </h2>

```python
import math


#initialize scans array
scans = []

#extract features from each image and append to scans array
for scan in data:
    filtered_scan = inception.predict(np.expand_dims(scan, axis=0))
    scans.append(filtered_scan.flatten())
    
    
```

<h2>Set up Grid Search for Cluster Size and Iteration Selection</h2>

```python
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.model_selection import GridSearchCV

#set hyperparameters
hyper_parameters={'n_clusters':[2,4,8,16],'max_iter':[300,600,900]}

#create kmeans instance
Kmeans = MiniBatchKMeans(verbose=1) 

#set up grid search
model=GridSearchCV(Kmeans,hyper_parameters,cv=10)

```

<h2>Create Soft Labels from Cluster Scans</h2>

```python
import matplotlib.pyplot as plt
import numpy as np

#generate clusters from scans
model.fit(scans)

#predict clusters of scans
labels = model.predict(scans)

```

<h2>Save Soft Labels to Transition Metadata</h2>

```python
import pandas as pd
# os.chdir('Images-processed')
df = pd.read_csv(os.path.join(os.getcwd(),'transition_metadata.csv'))

#prepend Soft labels with filler data for corrputed slices
filler = [99]*424

#load soft labels
df['Soft Labels'] = filler + list(labels)
df.to_csv(os.path.join(os.getcwd(),'transition_metadata.csv'))
```
